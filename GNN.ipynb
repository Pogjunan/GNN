## `-` GNN

`Graph` 의 구성요소 <br>
`-` $Graph = G(V,E)$
<br>
Vertices = 화살표 <br> 
nodes = 동그라미 <br>
Edges = 이어진 


Vertices = input 이고 각각의 feature <br>
Edges = input 을 이어주며 방향이 있을 수 도없을 수 도 있다.
